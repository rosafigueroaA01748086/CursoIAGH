# Matriz de confusión

**Matriz de confusion**

Objetivo: Evalua el rendimiento de un modelo de clasificación

Columnas: Valores predichos
Filas: Valores reales

* Verdadero Positivo (TP): Prediccion es correctas para la clase positiva

* Falso Positivo (FP): Predicciones incorrectas de la clase positiva

* Verdadero Negativo (TN): Predicciones correctas para la clase negativa

* Falso Negativo (FP): Predicciones incorrectas de la clase negativa


La matriz ayuda a calcular otras métricas para evaluar el modelo:

**Exactitud (accuracy)**: Cantidad de predicciones que acertó el modelo.

* **Accuracy= (TP+TN)/(TP+TN+FP+FN)**


**Precisión (precision)**: Se centra en la calidad de las predicciones positivas del modelo. Nos indica cuantas predicciones positivas fueron correctas.  

* **Precision= TP/(TP+FP)**

**Recuperación (recall)**: Mide la eficacia del modelo para predecir casos positivos. Muestra la proporción de TP detectados con respecto a todos los casos reales positivos.

* **Recall= TP/(TP+FN)**

**Especificidad** : Mide la capacidad de identificar correctamente la clase negativa.

* **Specificity= TN/(TN+FP)**

Calcular las 4 métricas

```python
#Valores
TP = 412
TN = 2137
FP = 256
FN = 17

#Accuracy

acc = (TP+TN)/(TP+TN+FP+FN)

print(f'El valor de exactitud (accuracy) es: ', acc)
```

```python
# Precision

precision = TP/(TP+FP)

print(f'El valor de precisión (precision) es: ', precision)
```

```python
# Recall - Sensibilidad

recall = TP/(TP+FN)

print(f'El valor de sensibilidad (recall) es: ', recall)
```

```python
# Especifidad

specificity = TN/(TN+FP)

print(f'El valor de especifidad (specificity) es: ', specificity)
```
